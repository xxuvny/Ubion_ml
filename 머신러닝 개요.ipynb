# **머신러닝 기초**

### **1. 머신러닝이란?**  

https://blog.naver.com/tommybee/222639011782   
   
    
특수하게 프로그래밍되어 '스스로 학습하고 개선하는 시스템' 을 만드는 프로세스   

---
**1.1. 머신러닝 목표**
 - 시스템이 자동으로 데이터를 수집하고 해당 데이터를 사용하여 더 많은 것을 학습하는 데 도움이 되는 알고리즘을 설계하는 것   
 
 - 시스템은 수집된 데이터에서 패턴을 찾고 이를 사용하여 중요한 결정을 내려야 함   
     
 - 시스템이 인간처럼 생각하고, 행동하고, 인간과 같은 지능을 보여주고 생각할 수 있는 능력을 제공함

<br>

**1.2. 머신러닝 모델 예시**

* 현재 모델  
> G-mail 등의 스팸 메일 분리    
> 자동 고침에서 볼 수 있는 문법 및 철자 오류 수정  

<br>

* 발전 중인 모델
> 사물 및 이미지 인식   
> 가짜 뉴스 탐지   
> 글(문장)이나 말 이해하기
> 사람과 상호작용하는 웹사이트의 봇(챗봇 등)
> 자율 주행 자동차..... 등

<br>

**1.3. 단계**

<u>(1) 데이터 수집</u>
> 모델의 정확성 = 기계에 제공하는 데이터의 품질에서 결정
> 정확하지 않거나 오래된 데이터가 있는 경우, 관련성이 없는 잘못된 결과 또는 예측 발생   
   

<u>(2) 데이터 전처리</u>   
> 데이터의 '무작위화' : 데이터가 고르게 분포되고 순서가 학습 프로세스에 영향을 미치지 않도록 하는 데 도움   
> 데이터 정리    
      * 제거 : 불필요한 데이터, 누락된 값, 중복값 등      
      * 데이터 유형 변환   
      * 데이터 집합 재구성   
      * 행/열 재구성 또는 행/열 인덱스 변경 등   
> 데이터 시각화 : 데이터 방식, 변수 등을 파악 또는 이해하기 위해 시행   
> 데이터 분리 : 2개의 데이터 집합으로 분리   
      * 훈련 집합 : 모델이 학습하는 데이터 집합 (dataset)   
      * 테스트 집합 : 훈련 후 모델의 정확도를 확인하기 위해 사용   
   

<u>(3) 모델 선택</u>
> 음성 인식, 이미지 인식, 예측 등과 같은 적절한 모델을 선택 또는 개발   
> 당면한 작업과 관련된 모델을 선택하는 것이 중요   


<u>(4) 모델 훈련</u>
> 머신러닝에서 가장 중요한 단계  
> 준비된 데이터를 머신러닝 모델에 전달하여 패턴을 찾고 예측 
> 데이터에서의 학습을 반복하면서 시간이 지남에 따라 더 나은 예측을 하게 됨


<u>(5) 모델 평가</u>
> 모델 훈련시킨 후 성능을 확인하는 과정   
> 처음에는 분할해 두었던 test 용 데이터 집합을 사용함   


<u>(6) 매개변수 조정</u>
> 모델 평가 후 정확도를 개선할 수 있는지 확인하는 과정 : 매개변수를 조정하여 수행   
   > 매개변수 = 프로그래머가 결정하는 모델의 변수
> 특정 매개변수 값에서 정확도가 최대값이 되므로 이 값을 찾는 것이 중요


<u>(7) 예측</u>
> 머신러닝 모델 수행

### **3. 머신러닝 유형**  

https://blog.naver.com/tommybee/222639032336  
   
     
**3.1. 지도 학습**
 - 레이블이 지정된 데이터를 사용하여 머신러닝 모델을 훈련하는 유형   
 - 일반적으로 분류 및 회귀 문제를 해결하는 데 사용
    
 - 지도학습의 주요 알고리즘
   > 선형 회귀          (Linear Regression)   
   > 로지스틱 회귀      (Losistic Regression)   
   > 서포트 벡터 머신   (Support Vector Machine)   
   > K-최근접 이웃법    (KNN, K Nearest Neighbor)   
   > 의사 결정 나무     (Decision Tree)   
   > 랜덤 포레스트      (Random Forest)   
   > 나이브 베이즈 분류 (Naive Bayes)   
   
 - 지도학습 활용 예시
   > 날씨 예측   
   > 판매 예측   
   > 주가 분석   

<br>

**3.2. 비지도 학습**
 - 레이블이 지정되지 않은 데이터를 사용하여 기계를 훈련하는 머신러닝 유형
 - 스스로 패턴을 찾고 데이터의 추세를 이해하여 결과를 발견 → 모델 구축 시 감독이 필요없음   
 - 주로 클러스터링(그룹화) 또는 연관 문제를 해결하는 데 사용   
   
 - 알고리즘 : 해결하려는 문제 유형에 따라 다름
 - 비지도 학습의 주요 알고리즘
    > <u> K Means 클러스터링 </u>  
    > 계층적 클러스터링   (Hierarchical Clustering)   
    > DBSCAN   
    > 주 성분 분석        (Principal Component Analysis)   

 - 비지도학습 활용 예시
    > 고객 세분화 : 고객 행동, 선호도 및 관심사를 기반으로 유사한 고객을 그룹으로 분류하고 클러스터링 가능   
    > 이탈률 분석

<br>

**3.3. 강화 학습**
 - 기계가 특정 상황에 적절한 조치를 취하고 보상을 극대화하도록 훈련
    - 에이전트와 환경을 사용하여 작업과 보상을 생성
    - e.g. 개에게 공을 잡도록 훈련시키는 과정 : 성공하면 비스킷과 같은 보상을 줌
 - 미리 정의된 대상 변수 없음

 - 강화학습의 주요 알고리즘
    > Q-learning
    > SARSA
    > Monte Carlo
    > Deep Q network

 - 강화학습 활용 예시
    > 게임 산업 : 게임을 구축하는 데 널리 사용
    > 로봇 훈련

### **4. 지도 학습**

**4.1. 지도 학습의 두가지 형태**
 * 분류 Classification   
 * 회귀 Regression   


**4.2. Classification**
 - 출력 변수가 [ 범주형 데이터 ] 일 때 사용
 - 범주형 데이터 : 예/아니오, 남성/여성, 참/거짓.. 등
 - e.g. 스팸메일 분류


**4.3. Regression**
 - 출력 변수가 [ 실수 또는 연속값 ] 일 때 사용   
 - 둘 이상의 변수 사이에서 연관 관계를 가짐   
     - 예를 들면, 경력에 따른 급여, 키에 따른 체중.. 등   
     - 한 변수가 변경되면 다른 변수에도 영향을 끼침
    
 <u>e.g. 온도와 습도</u>   
   > 온도 : 독립 변수 | 습도 : 종속 변수   
   > 온도가 높아지면 습도가 낮아짐   
     > 온도와 습도의 상관관계가 모델에 제공되고, 기계는 이 관계를 학습함     
     > 훈련 후 주어진 온도를 기반으로 습도 예측     

<br>

**4.4. 지도 학습의 실제 적용**

 (1) 위험성 평가 (Risk Assesment)   
      - 회사의 위험 포트폴리오 최소화   
      - 금융서비스 또는 보험영역의 위험 평가에 사용
<br>

 (2) 이미지 분류 (Image Classification)   
      - 지도형 머신러닝을 시연하는 주요 사용 사례 중 하나   
      - Facebook 예시 : 태그가 지정된 사진앨범의 사진에서 친구를 인식할 수 있음   
<br>

 (3) 사기 탐지 (Fraud Detection)   
      - 사용자 거래의 진위 여부 식별   
<br>    

 (4)시각적 인식 (Visual Recognition)   
      - 사물, 장소, 사람, 행동, 이미지 등 식별

### **5. 비지도 학습**

레이블이 지정되지 않은 데이터를 사용하여 감독 없이 스스로 학습    

**5.1. 비지도 학습의 두가지 형태**
 * 군집화 Clustering   
 * 연관   Association   


**5.2. Clustering**
 - cluster : 사전적 의미 - 통계 조사의 대상인 모집단의 요소를 몇 개 모은 단위체 
 - 객체를 유사한 클러스터로 나누고 다른 클러스터에 속하는 객체와 유사하지 않은 클러스터로 나누는 방법   


**5.3. 연관**
 - 선택 항목 중 '동시에 발생할 확률'을 발견하는 규칙 기반 머신 러닝


**5.4. 비지도 학습의 실제 적용**

 (1) 시장 바구니 분석   
 (2) 시맨틱 클러스터링 (Semantic Clustering)   
 (3) 배달 매장 최적화 : 수요 예측을 통해 공급 최적화   
 (4) 사고 취약 지역 파악

# **주요 용어 정의**

### **1. Feature**

**(1) 개념**

https://blog.naver.com/tommybee/222651276783  

 * '머신러닝 모델에 제공하는 입력변수'를 의미 

 * 피처 선택이 중요한 이유
   * 최적의 모델을 훈련시키기 위해 필수/주요 특성만 사용 필요
   * why? 특성이 너무 많으면 모델이 중요하지 않은 패턴까지 캡처하게 되므로 노이즈로부터도 학습할 위험이 있음
      * *"쓰레기(데이터 노이즈)를 넣으면 쓰레기가 출력된다"*

<br>

**(2) Feature Selection** (특성 선택)

**I. What is?**

* 관련 데이터만 사용하여 모델에 대한 입력변수를 줄이고 데이터의 노이즈를 제거하는 방법  
  * 즉, 데이터의 중요한 매개변수를 선택하는 방법
  * 더 빠르고 정확한 모델을 만들기 위해서는 올바른 데이터를 선택해야만 한다.
 <br> 
 <br>   
        
**II. 피처 선택 모델 유형**

* 감독 모델
  * 특성 선택을 위해 출력 레이블 클래스를 사용하는 방법
  * 목표 변수들이 모델의 효율성을 높일 수 있는 변수들을 식별하는 데 사용  
<br>  

* 비지도 모델
  * 특성 선택을 위해 레이블 클래스가 필요하지 않은 방법
  * 레이블이 지정되지 않은 데이터에 사용함

### **2. y = f(X)**


**1. Label**
 * 함수식에서 'y' 변수에 해당하는 값

 * 학습 완료된 모델의 경우, '기계가 도출해 주는 정보' 임

 * 예 : 농작물의 향후 예측 가격, 사진에 표시되는 동물의 종류, 동영상의 의미 등 무엇이든 가능

<br>

**2. Sample**
  * 함수식에서 'X' 에 해당하는 값 = 머신러닝 모델에 제공하는 feature 에 해당

  * 샘플은 기계 학습에 주어지는 특정한 예제를 의미함

  * 학습 완료된 모델의 경우, '기계에게 입력해 주어야 하는 정보' 임

### **예측** (Prediction)

 * 즉, 학습된 모델을 사용하여 유용한 예측(y')을 해내는 것
